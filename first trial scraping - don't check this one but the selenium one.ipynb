# Parsing it (Soma's way)

```python
response = requests.get(urltime)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
#let's look at the doc
#it should give 25 hotels, including 'GINN hotel'
#however, it's not showing up. What's wrong?
```

```python
doc
```

```python
names = doc.find_all('span', itemprop='name')
```

```python
prices = doc.find_all(class_='bui-price-display__value')
```

```python
for price, name in zip(prices, names):
    print('---')
    print(name.text)
    print(price.text)
```

```python
rows = []

for price, name in zip(prices, names):
    row = {}
    print('---')
    #print(name.text)
    #print(price.text)
    row['Hotel'] = name.text
    row['Price'] = price.text
    print(row)
    rows.append(row)
```

```python
df = pd.DataFrame(rows)
df
```

```python
df=df.replace(to_replace = '€', value ='')
df
```
